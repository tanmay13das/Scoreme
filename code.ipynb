```python
import os
import pandas as pd
from PyPDF2 import PdfReader
from openpyxl import Workbook

# Specify the PDF file path
pdf_file = "test3 (1).pdf"

# Check if the file exists
if not os.path.exists(pdf_file):
    print(f"Error: File '{pdf_file}' does not exist.")
else:
    try:
        # Read the PDF file
        reader = PdfReader(pdf_file)
        
        # Initialize a list to store extracted table data
        extracted_data = []

        # Process each page
        for page in reader.pages:
            text = page.extract_text()

            # Skip empty pages
            if not text.strip():
                continue

            # Split the text into lines
            lines = text.splitlines()

            # Detect and process table-like content
            for line in lines:
                print(line)
                # Example detection logic for tabular data
                if any(char.isalnum() for char in line):
                    extracted_data.append(line.split())

        # Convert the extracted data into a DataFrame
        df = pd.DataFrame(extracted_data)

        # Save the DataFrame to an Excel file
        output_file = "extracted_tables.xlsx"
        with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
            df.to_excel(writer, index=False, header=False, sheet_name="Extracted Tables")

        print(f"Tables extracted and saved to '{output_file}'.")

    except Exception as e:
        print(f"An error occurred while processing the PDF: {e}")
```
